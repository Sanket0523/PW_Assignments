Q-What is App routing in Flask? Why do we use app routes?

In Flask, App routing is the process of defining URLs (Uniform Resource Locators) that the application should respond to. When a user makes a request to a specific URL, the application matches that URL to a corresponding view function that generates a response.

We use app routes in Flask to define the behavior of the web application based on the user's requests. App routes help map URLs to specific functions in the application, so when a user enters a URL into their browser, Flask knows which function to call to handle that request.


```python
from flask import Flask

app = Flask(__name__)

@app.route("/hello")
def hello():
    return "Hello World!"

if __name__ == "__main__":
    app.run(host="0.0.0.0")
```

In this example, the @app.route('/') decorator is used to define a route for the root URL (/). The index() function is called when a user navigates to the root URL, and it returns the string "Hello, World!" as the response.

App routing is a fundamental feature of Flask, and it is important to understand how it works when building Flask applications. With app routes, we can define the behavior of our application and create a RESTful API by mapping HTTP methods (GET, POST, PUT, DELETE) to functions that handle those requests.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.


In Flask, the url_for() function is used for URL building. It generates a URL to a specific function. This function takes the name of the function as the first argument and any number of keyword arguments, each corresponding to a variable part of the URL rule.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'This is the home page'

@app.route('/about')
def about():
    return 'This is the about page'

@app.route('/contact')
def contact():
    return 'This is the contact page'

@app.route('/user/<username>')
def user_profile(username):
    return f'This is {username}\'s profile page'

with app.test_request_context():
    print(url_for('home'))  # Output: /
    print(url_for('about'))  # Output: /about
    print(url_for('contact'))  # Output: /contact
    print(url_for('user_profile', username='john'))  # Output: /user/john

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
