```python
def test():
    pw=input("enter the pw: ")
    l1=[]
    l2=[]
    l3=[]
    l4=[]
    if len(pw) >= 10:
        for i in pw:
            if i.isupper():
                l1.append(i)
            elif i.islower():
                l2.append(i)
            elif i.isdigit():
                l3.append(i)
            else:
                l4.append(i)
        if len(l1)>1 and len(l2)>1 and len(l3)>0 and len(l4)>2:
            print("Valid password")
        else:
            print("Invalid password")
    
    else:
        print("Invalid password")
test()
```

```python
#sort based on key:
l = [("mango", 99), ("orange", 80), ("grapes", 1000)]
l.sort(key=lambda i: i[1])
print(l)
```

OR

```python
#sort based on values:
l = [("mango", 99), ("orange", 80), ("grapes", 1000)]
l.sort(key=lambda i: i[0])
print(l)
```
