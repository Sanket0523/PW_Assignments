```python
from functools import reduce
```

```python
list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,22, 61, 34)}, 
         [56, 'data science'], 'Machine Learning']
l=[]
product=1
for i in list1:
    if type(i)==int:
        l.append(i)
    elif type(i)==list:
        for j in i:
            if type(j)==int:
                l.append(j)
    elif type(i)==tuple:
        for k in i:
            if type(k)==int:
                l.append(k)
    elif type(i)==set:
        for m in i:
            if type(m)==int:
                l.append(m)
    elif type(i)==dict:
        for n in i.keys():
            if type(n)==int:
                l.append(n)
        for o in i.values():
            if type(o)==int:
                l.append(o) 
            elif type(o)==list :
                for q in o: 
                    l.append(q)
            elif type(o)==tuple :
                for r in o:
                    l.append(r)
l
reduce(lambda i,j : i*j , l)
```
