```python
If we want to execute a group of statements multiple times then we should 
go for Iterative statements. 
Python supports 2 types of iterative statements. 
1) for loop 
2) while loop

If we want to execute some action for every element present in some sequence 
(it may be string or collection) then we should go for for loop.

we want to execute a group of statements iteratively until some condition false,
then we should go for while loop.
```

```python
#for loop example
l=[1,2,3]
for i in l:    #execute saction for every element present
    print(i)
```

```python
#while loop example
n=int(input("enter the number"))
i=1
while i<=n:    #execute action until some condition
      print(i)
      i=i+1
```

```python
#python program to print sum and product of first 10 natural number using for loop

sum=0
product=1
for i in range(1,11):
    sum=sum+i
    product=product*i
print("The sum of first 10 natural number is: ",sum)
print("The Product of first 10 natural number is: ",product)
```

```python
#python program to print sum and product of first 10 natural number using while loop
i=1
sum=0
product=1
while i<=10:
    sum=sum+i
    product=product*i
    i=i+1
print("The sum of first 10 natural number is: ",sum)
print("The Product of first 10 natural number is: ",product)
```

```python
n=int(input("Enter the units of electricity consumed in a month:"))
if n<=100:
    print("Total Electricity bill is :",n*4.5)
elif n>100 and n<=200:
    print("Total Electricity bill is :",450+(n-100)*6)
elif n>200 and n<=300:
    print("Total Electricity bill is :",450+600+(n-200)*10)
elif n>300:
    print("Total Electricity bill is :",450+600+1000+(n-300)*20)
```
